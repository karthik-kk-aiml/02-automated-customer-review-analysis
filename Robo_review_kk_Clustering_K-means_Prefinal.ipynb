```python
df = pd.read_csv(r"dataset/1429_1.csv")
df = df.dropna(subset=["reviews.rating"])
# Add a 'label' column based on 'reviews.rating' which will indicate if it's positive or negative
def label_sentiment(rating):
    if rating in [1, 2]:
        return 0  # Negative
    elif rating == 3:
        return 1  # Neutral
    else:
        return 2  # Positive


# Apply the function to create a new column
df["label"] = df["reviews.rating"].apply(label_sentiment)

# Drop rows where 'reviews.text' is missing (if any)
df = df.dropna(subset=["reviews.text"])
```

```python
# Step 1: Load the Dataset
#df_loaded = pd.read_pickle("df1429_1.pkl")
#df = df_loaded

# Inspect the dataset
print(df.shape)
print(df.head())
print(df.isnull().sum())
```

```python
# Drop rows where following values are missing (if any)
df = df.dropna(
    subset=["reviews.text", "name", "categories", "reviews.rating", "reviews.title"]
)  # Ensure no missing data
df = df.drop(
    [
        "reviews.dateAdded",
        "reviews.didPurchase",
        "reviews.id",
        "reviews.userCity",
        "reviews.userProvince",
    ],
    axis=1,
)

print(df.shape)
print(df.columns)
print(df.isnull().sum())
```

```python
# Convert 'name' and 'categories' fields into numercial feature that K-Means can understand

from sklearn.feature_extraction.text import TfidfVectorizer

# Combine the 'name' and 'categories' fields into one text feature
df["combined"] = df["name"] + " " + df["categories"]

# Use TfidfVectorizer to convert the text data into numerical features
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df["combined"])
```

```python
from sklearn.cluster import KMeans

# Apply K-Means clustering with 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
df["category_label"] = kmeans.fit_predict(X)
```

```python
print(df["category_label"].value_counts())
```

```python
df.to_csv("categorized_dataset_k5.csv", index=False)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce the dimensionality of the TF-IDF features to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())  # Convert sparse matrix to dense array for PCA

# Create a scatter plot of the clusters
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    X_pca[:, 0], X_pca[:, 1], c=df["category_label"], cmap="viridis", alpha=0.6
)

# Assuming there are 5 clusters, for example
cluster_names = {
    0: "Fire Tablet 7",
    1: "Kindle",
    2: "Speakers/Streaming",
    3: "Fire HD 8",
    4: "Fire KIDS",
}

# Add a legend with the meaningful category names
plt.legend(handles=scatter.legend_elements()[0], labels=list(cluster_names.values()))
plt.title("K-Means Clustering of Product Categories")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

```python
# Plotting scatterplot with cluster names and also saving it to a file!

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd

# Reduce the dimensionality of the TF-IDF features to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray())  # Convert sparse matrix to dense array for PCA

# Create a scatter plot of the clusters with contrasting colors using 'tab10' colormap
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=df["category_label"],
    cmap=plt.cm.get_cmap("tab10", 6),
    alpha=0.6,
)

# Assuming you have 5 or 6 clusters with meaningful names
cluster_names = {
    0: "Fire Tablet 7",
    1: "Kindle",
    2: "Speakers/Streaming",
    3: "Fire HD 8",
    4: "Fire KIDS",
}

# Add the meaningful cluster names to the DataFrame as a new column
df["category_name"] = df["category_label"].map(cluster_names)

# Add a legend with the meaningful category names
plt.legend(handles=scatter.legend_elements()[0], labels=list(cluster_names.values()))
plt.title("K-Means Clustering of Product Categories")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()

# Save the DataFrame to a CSV file with the new 'category_name' column
df.to_csv("categorized_dataset_k5_with_names.csv", index=False)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

inertia = []
k_values = range(1, 11)  # Choose a range for the number of clusters

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Diagram
plt.figure(figsize=(8, 6))
plt.plot(k_values, inertia, 'bo-', color='b')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Diagram')
plt.show()
```
